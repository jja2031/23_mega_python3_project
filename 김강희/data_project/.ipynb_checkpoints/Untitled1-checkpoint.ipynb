```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df
```

```python
countryList = ['European Union', 'Nigeria', 'Iran']
temp = []

for countryName in countryList:
    temp.append(df[df['category'] == countryName])

res = pd.concat(temp)
res
```

```python
years = ['2020', '2021']
temp = res[res['category'] == 'Iran']
temp1 = temp[years]
temp2 = list(temp1.values)
sum(temp2[0])/len(temp2[0])
```

```python
countryList =  ['European Union', 'Nigeria', 'Iran']
years = ['2020', '2021']
res = []
for countryName in countryList:
    temp = (df[df['category'] == countryName])
    temp1 = temp[years]
    temp2 = list(temp1.values)
    result = sum(temp2[0])
    res.append(result)
res
```

```python
x = countryList
y = res
plt.bar(x, y)
plt.show()
```

```python
countryList = ['Japan','South Korea']
years = ['1990','1991','1992','1993','1994']
res = []
for countryName in countryList:
    temp = (df[df['category'] == countryName])
    temp1 = temp[years]
    temp2 = list(temp1.values)
    result = sum(temp2[0])
    res.append(result)
    
x = countryList
y = res
plt.bar(x, y)
plt.show()
```

```python
local = "Europe"
local2 = "South Korea"
standardYear = 1990
tempDf = df[df['category'] == local]
tempdf = df[df['category'] == local2]
```

```python
transposeDf = tempDf.transpose()
tempList = transposeDf.values.tolist()
resList = []
for item in tempList:
    resList.append(item[0])
resList = resList[1:-2]
x = list(range(len(resList)))
for i in range(len(x)):
    x[i] += standardYear
y = resList

transposedf =  tempdf.transpose()
templist = transposedf.values.tolist()
reslist = []
for item in templist:
    reslist.append(item[0])
reslist = reslist[1:-2]
z = list(range(len(reslist)))
for i in range(len(z)):
    z[i] += standardYear
c = reslist
```

```python
plt.plot(x, y, z, c, 'r-')
plt.show()
```

```python
localList = ["World",'OECD','G7','BRICS']
standardYear = 1990
tempDf = []

for i in range(len(localList)):
    transDf = df[df['category'] == localList[i]].transpose().values.tolist()
    resList = []
    for item in transDf:
        resList.append(item[0])
    resList = resList[1:-2]
    tempDf.append(resList)
```

```python
x = list(range(len(tempDf[0])))
for i in range(len(x)):
    x[i] += standardYear
for i in range(len(tempDf)): 
    y = tempDf[i]
    plt.plot(x, y)

plt.show()
```
